```python
import sys, platform, subprocess

print("=== Python & OS ===")
print(f"Python: {sys.version.split()[0]}")
print(f"Platform: {platform.platform()}")

print("\n=== CPU ===")
cpuinfo = subprocess.run(["cat", "/proc/cpuinfo"], capture_output=True, text=True).stdout
model = [l for l in cpuinfo.split("\n") if "model name" in l]
cores = cpuinfo.count("processor\t:")
print(f"Logical CPUs: {cores}")
if model:
    print(f"Model: {model[0].split(':')[1].strip()}")

print("\n=== RAM ===")
print(subprocess.run(["free", "-h"], capture_output=True, text=True).stdout)

print("=== Disk (/content) ===")
print(subprocess.run(["df", "-h", "/content"], capture_output=True, text=True).stdout)

print("=== GPU ===")
gpu = subprocess.run(["nvidia-smi", "--query-gpu=name,memory.total,driver_version", "--format=csv"], capture_output=True, text=True)
print(gpu.stdout if gpu.returncode == 0 else "No GPU detected (will use CPU)")

print("=== Key ML libraries ===")
for pkg in ["numpy", "pandas", "sklearn", "xgboost", "matplotlib", "seaborn"]:
    try:
        m = __import__(pkg)
        print(f"  {pkg}: {m.__version__}")
    except Exception:
        print(f"  {pkg}: NOT INSTALLED")
```
