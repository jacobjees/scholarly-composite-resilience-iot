# 01 — Mount Drive and create project folder structure

Run once at the start of a session. Sets up the canonical project layout under
`MyDrive/composite_resilience_framework/`. Idempotent — safe to re-run.

```python
import os
from pathlib import Path
from google.colab import drive

DRIVE_MOUNT = '/content/drive'
if not os.path.ismount(DRIVE_MOUNT):
    drive.mount(DRIVE_MOUNT)
    print(f'Drive mounted at {DRIVE_MOUNT}')
else:
    print(f'Drive already mounted at {DRIVE_MOUNT}')

PROJECT_ROOT = '/content/drive/MyDrive/composite_resilience_framework'
SUBDIRS = [
    'data/raw',         # the CICIoT2023 CSV.zip lives here (you place it manually)
    'data/processed',   # train/val/test parquets — generated by notebook 04
    'notebooks',
    'results',          # metrics, summaries, LaTeX tables
    'figures',          # paper-ready PNG and PDF
    'models',           # trained model checkpoints
    'logs',
]

print(f'\nCreating project folders under: {PROJECT_ROOT}')
for s in SUBDIRS:
    full = os.path.join(PROJECT_ROOT, s)
    os.makedirs(full, exist_ok=True)
    print(f'  ok  {s}/')

print('\nFolder tree:')
for root, dirs, files in os.walk(PROJECT_ROOT):
    depth = root.replace(PROJECT_ROOT, '').count(os.sep)
    indent = '  ' * depth
    print(f'{indent}{os.path.basename(root)}/')

print('\nSetup complete. Next: notebook 02 (extract CSV.zip and verify).')
```
